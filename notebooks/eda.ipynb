```python
df_4th = df[
    (df["down"] == 4) &
    (df["play_type"].isin(["run", "pass", "punt", "field_goal"])) &
    (df["season_type"] == "REG") &
    (df["epa"].notnull()) &
    (df["yardline_100"].notnull()) &
    (df["ydstogo"].notnull())
]

```

```python
columns_to_keep = [
    "game_id",
    "play_id",
    "season",
    "week",
    "game_date",
    "qtr",
    "posteam",
    "defteam",
    "down",
    "ydstogo",
    "yardline_100",
    "goal_to_go",
    "play_type",
    "game_seconds_remaining",
    "half_seconds_remaining",
    "quarter_seconds_remaining",
    "game_half",
    "score_differential",
    "yards_gained",
    "epa",
    "success",
    "play_type",
    "posteam_timeouts_remaining",
    "defteam_timeouts_remaining",
    "wp",
    "def_wp",
    "home_wp",
    "away_wp",
    "wpa",
    "field_goal_result",
    "kick_distance",
    "punt_blocked",
    "no_huddle",
    "shotgun",
    "qb_dropback",
    "desc",
]
```

```python
df_4th = df_4th[columns_to_keep]
```

```python
import matplotlib.pyplot as plt
```

```python
print(df_4th.columns[df_4th.columns.duplicated()])
```

```python
df_4th = df_4th.loc[:, ~df_4th.columns.duplicated()]
```

```python
# Only plays where team goes for it on 4th
conversion_attemps = df_4th[df_4th["play_type"].isin(["run", "pass"])]

# Clean and Sort
conversion_rate = (
    conversion_attemps.groupby("ydstogo")["success"]
    .mean()
    .reset_index()
    .sort_values("ydstogo")
)

# Casting yds to go as integer
conversion_rate["ydstogo"] = conversion_rate["ydstogo"].astype(int)

# Plot with matplotlib
plt.figure(figsize=(10, 6))
plt.bar(conversion_rate["ydstogo"], conversion_rate["success"])
plt.title("4th Down Conversion Rate by Yards to Go")
plt.xlabel("Yards to Go")
plt.ylabel("Conversion Rate")
plt.xticks(conversion_rate["ydstogo"])
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
epa_by_type = df_4th.groupby("play_type")["epa"].mean()
epa_by_type.plot(kind="bar", title="Average EPA by 4th Down Play Type")
```

```python
# Binning Field Position into 10-yard chunks
df_4th["yard_bin"] = pd.cut(df_4th["yardline_100"], bins=range(0, 110, 10))

#Comparing average EPA by type and position on field
epa_by_position = df_4th.groupby(["yard_bin", "play_type"])["epa"].mean().unstack()

# Visualizing
epa_by_position.plot(kind="line", title="EPA by Field Position and Play Type")
```
