```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:\\Users\\jseidl\\NFL-4th-Down\\data\\processed\\eval_results.csv")
```

```python
df = df[df['actual'].isin(['Go', 'No Go'])]
df = df[df['recommended'].isin(['Go', 'No Go'])]
```

```python
match_counts = df['matched'].value_counts()

plt.figure(figsize=(5, 5))
plt.pie(match_counts, labels=["Matched", "Did Not Match"], autopct='%1.1f%%', startangle=90)
plt.title("Tool Recommendation Match Rate")
plt.tight_layout()
plt.show()
```

```python
success_rates = {
    "Success When Matched": df[df['matched']]['success'].mean(),
    "Overall Success": df['success'].mean()
}

plt.figure(figsize=(6, 4))
sns.barplot(x=list(success_rates.keys()), y=list(success_rates.values()))
plt.ylabel("Success Rate")
plt.title("Success Rate Comparison")
plt.ylim(0, 1)
plt.tight_layout()
plt.show()
```
